```python
import json, sys, pathlib

sys.path.insert(0, str(pathlib.Path("..").resolve()))

from document_parser import HybridDocumentParser

parser = HybridDocumentParser(debug=True)
```

```python
# ── 测试1：parse 478.pdf（PDF文件）──
pdf_path = "../../test_data/478.pdf"
result_pdf = parser.parse(pdf_path)

print(f"文件名: {result_pdf.file_name}")
print(f"文件类型: {result_pdf.file_type}")
print(f"sections 数量: {len(result_pdf.sections)}")
print()
for i, sec in enumerate(result_pdf.sections):
    content_preview = str(sec.content)[:80].replace("\n", " ")
    print(f"[{i}] type={sec.type:6s}  title={sec.title!r:10s}  content={content_preview!r}")
```
