````python
# ── Section extraction config ──────────────────────────────────────────────
CATEGORY_SECTIONS = {
    "General": {
        "top_level":   ["SUMMARY INFORMATION", "LEAD APPLICANT & RESEARCH TEAM"],
        "app_details": ["Scientific Abstract"],
        "truncate":    {"Scientific Abstract": 3000},
    },
    "Proposed research": {
        "top_level":   [],
        "app_details": ["Scientific Abstract", "Plain English Summary"],
        "truncate":    {"Scientific Abstract": 2500, "Plain English Summary": 2500},
    },
    "Training and development": {
        "top_level":   [],
        "app_details": ["Training & Development and Research Support"],
        "truncate":    {"Training & Development and Research Support": 4000},
    },
    "Sites and support": {
        "top_level":   ["SUMMARY INFORMATION", "LEAD APPLICANT & RESEARCH TEAM"],
        # Include Scientific Abstract for supervisor/org context
        "app_details": ["Scientific Abstract"],
        "truncate":    {"Scientific Abstract": 2500},
    },
    "Working with people and communities": {
        "top_level":   ["SUMMARY BUDGET"],
        "app_details": ["Working with People and Communities Summary", "Plain English Summary"],
        "truncate":    {
            "Working with People and Communities Summary": 4500,
            "SUMMARY BUDGET": 2000,
        },
    },
    "Application Form": {
        "top_level":   [],
        "app_details": ["Scientific Abstract", "Plain English Summary",
                        "Working with People and Communities Summary"],
        "truncate":    {"Scientific Abstract": 1500, "Plain English Summary": 1500,
                        "Working with People and Communities Summary": 1500},
    },
}


def _trunc(text, limit):
    if limit and len(text) > limit:
        return text[:limit] + "\n...[truncated]"
    return text


def build_section_text(doc, category):
    cfg = CATEGORY_SECTIONS[category]
    parts = []
    for key in cfg['top_level']:
        val = doc.get(key)
        if not val:
            continue
        text = json.dumps(val, ensure_ascii=False, indent=2) if isinstance(val, dict) else str(val)
        parts.append(f'=== {key} ===\n{_trunc(text, cfg["truncate"].get(key))}')
    app = doc.get('APPLICATION DETAILS', {})
    for key in cfg['app_details']:
        val = app.get(key, '')
        if not val:
            continue
        parts.append(f'=== {key} ===\n{_trunc(str(val), cfg["truncate"].get(key))}')
    return '\n\n'.join(parts)


# ── Score derivation ────────────────────────────────────────────────────────
# Computed deterministically by Python — NOT by the LLM.
# present=2, partial=1, absent=0  ->  normalised to 1-10

_STATUS_PTS = {'present': 2, 'partial': 1, 'absent': 0}


def derive_score(signals):
    if not signals:
        return 1
    total = sum(_STATUS_PTS.get(s.get('status', 'absent'), 0) for s in signals)
    max_pts = len(signals) * 2
    return max(1, round((total / max_pts) * 10))


# ── Flatten General (mixed dict/list rubric structure) ───────────────────────
def flatten_general(rubric_general):
    items = []
    for key, val in rubric_general.items():
        if key == 'Applicant':
            items.extend(val)
        else:
            items.append({'name': key.replace('_', ' ').title(),
                          'definition': '', 'signals': val})
    return items


# ── LLM call ────────────────────────────────────────────────────────────────
SYSTEM_PROMPT = (
    'You are an expert NIHR grant application reviewer.\n\n'
    'For each criterion, check whether each listed signal is present in the application text.\n\n'
    'Signal status:\n'
    '  present  - clearly evidenced with specific, concrete detail\n'
    '  partial  - mentioned but vague or incomplete\n'
    '  absent   - not found in the text\n\n'
    'evidence: a short direct quote (2 sentences max) from the text, or null if absent.\n'
    'reasoning: 1-2 sentence overall assessment of the criterion.\n\n'
    'Return ONLY valid JSON. No markdown fences, no preamble, no extra text.'
)


def _extract_json(msg):
    """Extract JSON from Ollama message dict.
    Handles think=True quirks: CoT may appear in 'content' or 'thinking'.
    Tries both fields and strips markdown fences if present."""
    content  = msg.get('content',  '').strip()
    thinking = msg.get('thinking', '').strip()

    def _try(raw):
        if not raw:
            return None
        # Strip markdown code fence
        if '```' in raw:
            for block in raw.split('```')[1::2]:
                b = block.lstrip('\n')
                if b.startswith('json'):
                    b = b[4:]
                s, e = b.find('{'), b.rfind('}') + 1
                if s != -1 and e > s:
                    try:
                        return json.loads(b[s:e])
                    except json.JSONDecodeError:
                        pass
        # Find outermost { }
        s = raw.find('{')
        if s == -1:
            return None
        e = raw.rfind('}') + 1
        if e <= s:
            return None
        try:
            return json.loads(raw[s:e])
        except json.JSONDecodeError:
            return None

    result = _try(content) or _try(thinking)
    if result is None:
        preview = (content or thinking)[:300]
        raise ValueError(f'No JSON object found. Preview: {preview!r}')
    return result


def score_category(category_name, criteria, section_text):
    schema_example = (
        '{\n'
        '  "criteria": [\n'
        '    {\n'
        '      "name": "criterion name",\n'
        '      "signals": [\n'
        '        {"signal": "signal text", "status": "present", "evidence": "quote"}\n'
        '      ],\n'
        '      "reasoning": "1-2 sentence summary"\n'
        '    }\n'
        '  ]\n'
        '}'
    )
    user_prompt = (
        '[CRITERIA TO EVALUATE]\n'
        + json.dumps(criteria, ensure_ascii=False, indent=2)
        + '\n\n[APPLICATION TEXT]\n'
        + section_text
        + '\n\nReturn JSON exactly like this:\n'
        + schema_example
    )
    resp = requests.post(
        f'{OLLAMA_BASE_URL}/api/chat',
        json={
            'model': MODEL_NAME,
            'messages': [
                {'role': 'system', 'content': SYSTEM_PROMPT},
                {'role': 'user',   'content': user_prompt},
            ],
            'options': {'temperature': 0.1, 'top_p': 0.9, 'num_predict': 8192},
            'think': True,
            'stream': False,
        },
        timeout=600,
    )
    data = _extract_json(resp.json()['message'])

    results = []
    for item in data.get('criteria', []):
        signals = item.get('signals', [])
        results.append({
            'category':  category_name,
            'name':      item.get('name', ''),
            'score':     derive_score(signals),
            'signals':   signals,
            'reasoning': item.get('reasoning', ''),
        })
    return results


print('Helpers loaded.')
````
